```python
import sys

import requests

# Configuration
URL = "http://localhost:5001/v1/convert/source"
payload = {
    "sources": [
        {
            "kind": "http",
            "url": "https://www.govinfo.gov/content/pkg/FR-2026-03-23/pdf/2026-05588.pdf",
        }
    ]
}

print("🔍 Step 1: Checking if the server is even reachable...")
try:
    # We use a short timeout here just to see if the port is open
    requests.get("http://localhost:5001/", timeout=2)
    print("✅ Server is reachable.")
except Exception:
    print("❌ Cannot connect to localhost:5001. Is your Docker container/service running?")
    sys.exit(1)

print("🚀 Step 2: Sending the PDF conversion request (this may take 30-60 seconds)...")
try:
    # Added a 120-second timeout because PDF OCR/Conversion is slow
    response = requests.post(URL, json=payload, timeout=120)

    print(f"📡 Server responded with Status Code: {response.status_code}")

    if response.status_code == 200:
        print("🎉 Success! Here is the data:")
        print(response.json())
    else:
        print(f"⚠️ Server returned an error: {response.text}")

except requests.exceptions.Timeout:
    print(
        "⏰ Error: The request timed out. The PDF might be too large or the server is overloaded."
    )
except Exception as e:
    print(f"🚨 An unexpected error occurred: {e}")
```
